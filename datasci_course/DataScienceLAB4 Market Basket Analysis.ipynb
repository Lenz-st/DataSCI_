```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
```

```python
data = pd.read_csv('basket_items.csv',sep=",")
data[:5]
```

```python
basket = data['items'].str.get_dummies(sep='|')
basket[:5]
```

```python
basket = basket.astype(pd.SparseDtype("bool"))
basket[:5]
```

```python
frequent_itemsets = fpgrowth(basket, min_support=0.01, use_colnames=True)
frequent_itemsets.sort_values(by='support', ascending=False)
```

```python
metric = "lift"
rules = association_rules(frequent_itemsets, metric=metric, min_threshold=1)
rules.sort_values('support', ascending=False, inplace=True)  # This returns None
rules
```
