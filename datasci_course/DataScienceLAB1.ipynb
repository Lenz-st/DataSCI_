```python
import pandas as pd
titanic = pd.read_csv('train.csv')
titanic
```

```python
titanic.describe()
```

```python
titanic.sort_values(by=['Pclass', 'PassengerId'], ascending=True,inplace=False)
```

```python
titanic.isnull().sum()
```

```python
titanic.Age.fillna('30',inplace=True)
titanic.Cabin.fillna('none',inplace=True)
titanic.Embarked.fillna('none',inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic['Age']=titanic['Age'].astype(int)
```

```python
titanic.info()
```

```python
number_survived = len(titanic[titanic['Survived']==0])
print("จำนวนคนเสียชีวิต = ",number_survived)
number_survived = len(titanic[titanic['Survived']==1])
print("จำนวนคนรอดชีวิต = ",number_survived)
```

```python
df_temp1 =  titanic[(titanic['Sex']=='female')& (titanic['Survived']==0)]
df_temp1
number_female_dead = df_temp1 .shape[0]
number_female_dead
```

```python
df_temp2 =  titanic[(titanic['Sex']=='male')& (titanic['Survived']==0)]
df_temp2
number_male_dead = df_temp2 .shape[0]
number_male_dead
```

```python
df_temp3 =  titanic[(titanic['Sex']=='female')& (titanic['Survived']==1)]
df_temp3
number_female_Survived = df_temp3 .shape[0]
number_female_Survived
```

```python
df_temp4 = titanic[(titanic["Sex"] == "male") & (titanic["Survived"] == 1)]
df_temp4
number_male_Survived = df_temp4 .shape[0]
number_male_Survived
```

```python
percentage_female_dead = number_female_dead * 100/len(titanic)
print("มีเพศหญิงเสียชีวิต {} คน".format(number_female_dead ))
print("มีเพศหญิงเสียชีวิตคิดเป็น {:.3f}%".format(percentage_female_dead))
percentage_male_dead = number_male_dead * 100/len(titanic)
print("มีเพศชายเสียชีวิต {} คน".format(number_male_dead ))
print("มีเพศชายเสียชีวิตคิดเป็น {:.3f}%".format(percentage_male_dead))
percentage_female_Survived = number_female_Survived * 100/len(titanic)
print("มีเพศหญิงรอดชีวิต {} คน".format(number_female_Survived ))
print("มีเพศหญิงรอดชีวิตคิดเป็น {:.3f}%".format(percentage_female_Survived))
percentage_male_Survived = number_male_Survived * 100/len(titanic)
print("มีเพศหญิงเสียชีวิต {} คน".format(number_male_Survived ))
print("มีเพศหญิงเสียชีวิตคิดเป็น {:.3f}%".format(percentage_male_Survived))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = titanic
sns.set(style='whitegrid')
def show_plot(plot_func, *args, **kwargs):
    plt.figure(figsize=(10,6))
    plot_func(*args, **kwargs)
    plt.show()
```

```python
survived_counts = titanic['Survived'].value_counts().reset_index()
survived_counts,columns = ["Survived","Count"]
survived_counts
```

```python
show_plot(sns.countplot, data=df , x='Survived')
```

```python
show_plot(sns.countplot, data=df, x='Sex', hue='Survived')
```

```python
show_plot(sns.histplot, df, x='Age', bins=30, kde=True)
```

```python
show_plot (sns.histplot, df, x='Age', bins=30, hue='Survived', multiple='stack')
```

```python
show_plot (sns.countplot, data=df, x='Pclass', hue='Survived')
```

```python
show_plot (sns.histplot, df ['Fare'].dropna(), bins=30, kde=True)

show_plot (sns.boxplot, data=df, x='Survived', y='Fare')
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
show_plot (sns.countplot, data=df, x='FamilySize', hue='Survived')
```

```python
show_plot (sns.countplot, data=df, x='Embarked', hue='Survived')

show_plot (sns.scatterplot,data=df, x='Age', y='Fare', hue='Survived')
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()

# 14

df['child'] = df['Age'] < 18
show_plot (sns.countplot, data=df, x='child', hue='Survived')
```

```python
# 15 การกระจายตัวของ SibSp bin=10
show_plot (sns.histplot, df, x='SibSp', bins=10, kde=True)
# 16 การกระจายตัวของ Parch bin=10
show_plot (sns.histplot, df, x='Parch', bins=10, kde=True)
```

```python
# 17 อัตราการรอดชีวิตของคนที่มากับญาติเทียบกับลำพัง
show_plot (sns.barplot, data=df, x='Sex', y='Survived', hue=df['FamilySize']>0)
```

```python
# 18 ความสัมพันธ์ระหว่าง Pclass และ ค่าโดยสาร (Fare)
show_plot (sns.boxplot, data=df, x='Pclass', y='Fare')
```

```python
# 19 อัตราการรอดชีวิตของผู้โดยสารที่ขึ้นเรื่อจากท่าเรือต่างจุดกัน
show_plot (sns.barplot, data=df, x='Embarked', y='Survived')
# 20 ความแตกต่างค่าโดยสารระหว่างเพศ
show_plot (sns.boxplot, data=df, x='Sex' , y='Fare')
```

```python
# 21 Pie Chart - สัดส่วนของผู้โดยสาร
plt.figure(figsize=(6,6))
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightpink'])
plt.title('Proportion of Passengers')
plt.show()
```

```python
# 22 KDE plot - การกระจายอายุของผุ้ัรอดชีวิตและไม่รอดชีวิต
plt.figure(figsize=(8,5))
sns.kdeplot(data=df[df['Survived']==1], x='Age', fill=True, label='Survived', color='blue', alpha=0.6)
sns.kdeplot(data=df[df['Survived']==0], x='Age', fill=True, label='Not Survived', color='red', alpha=0.6)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
# plt.legend()
plt.show()
```

```python
# 23 Count Plot - จำนวนผู้โดยสารตามท่าเรือขึ้นเรือ
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Embarked', palette='Set2')
plt.title('Number of Passengers by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# 24 Swarm Plot - การกระจายอายุในแต่ละชั้นที่นั่ง
plt.figure(figsize=(8,5))
sns.swarmplot(data=df, x='Pclass', y='Age', palette='Set3')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# 25 Pair Plot - ความสัมพันธ์ระหว่างตัวแปรเชิงตัวเลข
sns.pairplot(data=df, vars=['Age', 'Fare', 'Pclass'], hue='Survived', palette='hot',diag_kind='kde')
plt.show()
```

```python
# 26 Violin Plot - การกระจายอายุในแต่ละชั้นที่นั่ง
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x='Pclass', y='Age', palette='coolwarm',split=True)
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
# 27 Joint Plot - ความสัมพันธ์ระหว่างอายุและค่าโดยสาร
sns.jointplot(data=df, x='Age', y='Fare', hue ='Survived', kind='scatter', palette='coolwarm')
plt.show()
```

```python
# 28 Categorical Scatter Plot - อัตราการรอดชีวิตตามชั้นที่นั่งและเพศ
sns.catplot(data=df, x='Sex',y='Survived',col='Pclass', kind='bar',palette='hot')
plt.subplots_adjust(top=0.8)
plt.suptitle('Survival Rate by Passenger Class')
plt.show()
```
